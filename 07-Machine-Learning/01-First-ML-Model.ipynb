# Decision Tree Classifier

### Objective
Classify fruits based on **Weight** and **Color**.

### Dataset
```python
# Features = [Weight, Color]
features = [
    [150, 0],
    [170, 0],
    [130, 1],
    [120, 1]
]

labels = ["apple", "apple", "orange", "orange"]
```

### Workflow
```python
clf = tree.DecisionTreeClassifier()  # Create model
clf.fit(features, labels)            # Train model
clf.predict([[160, 0]])              # Predict
```

### Visualize Decision Tree
```python
tree.plot_tree(
    clf,
    feature_names=["Weight", "Color"],
    class_names=["apple", "orange"],
    filled=True,
    rounded=True
)
```

### Quick Revision
```python
DecisionTreeClassifier()  # Create model
fit(X, y)                 # Train model

```

**Color Encoding**
- `0` → Red
- `1` → Orange

````python
from sklearn import tree

# Training data: [Weight, Color]
# Color: 0 = Red, 1 = Orange
features = [
    [150, 0],
    [170, 0],
    [130, 1],
    [120, 1]
]

# Target labels
labels = ["apple", "apple", "orange", "orange"]

# Create and train the model
clf = tree.DecisionTreeClassifier()
clf.fit(features, labels)

# Predict new data
print("Prediction for [160, 0]:", clf.predict([[160, 0]])[0])
print("Prediction for [115, 1]:", clf.predict([[115, 1]])[0])
````

````python
from sklearn import tree
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(10, 6))

# Draw decision tree
tree.plot_tree(
    clf,
    feature_names=["Weight", "Color"],  # Input features
    class_names=["apple", "orange"],    # Output classes
    filled=True,                        # Color the nodes
    rounded=True,                       # Rounded boxes
    fontsize=12                         # Font size
)

plt.title("Decision Tree Classifier")
plt.show()
````
