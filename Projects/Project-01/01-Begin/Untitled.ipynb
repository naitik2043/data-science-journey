```python
import json
```

```python
# Lets write function to load the data
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("data.json")
```

```python
data
```

```python
# Function to display users and their connections
def display_users(data):
    print("=== Users and Their Connections ===\n")

    # Create a dictionary: user_id -> user_name
    user_dict = {user["id"]: user["name"] for user in data["users"]}

    for user in data["users"]:
        friend_names = [user_dict[fid] for fid in user["friends"]]

        liked_pages = []
        for page_id in user["liked_pages"]:
            for page in data["pages"]:
                if page["id"] == page_id:
                    liked_pages.append(page["name"])

        print(f"User: {user['name']}")
        print(f"Friends: {', '.join(friend_names)}")
        print(f"Liked Pages: {', '.join(liked_pages)}")
        print("-" * 40)

    print("\n=== Pages Information ===\n")

    for page in data["pages"]:
        print(f"Page ID: {page['id']}")
        print(f"Page Name: {page['name']}")
        print("-" * 30)

display_users(data)
```
