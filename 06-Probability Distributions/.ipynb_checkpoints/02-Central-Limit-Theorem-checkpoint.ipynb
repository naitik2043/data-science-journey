# Central Limit Theorem (CLT)

## Definition
The **Central Limit Theorem (CLT)** states that **the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large (typically \(n \ge 30\)), regardless of the population's original distribution.**

In simple words, if we repeatedly take random samples from any population and calculate their means, the distribution of those means will be approximately **normal**.

## Conditions
- Samples are selected randomly.
- Samples are independent.
- Sample size is sufficiently large (usually **n ≥ 30**).

## Formula

\[
\bar{X} \sim N\left(\mu,\frac{\sigma}{\sqrt{n}}\right)
\]

where:
- **μ** = Population Mean
- **σ** = Population Standard Deviation
- **n** = Sample Size
- **σ / √n** = Standard Error

## Complete Python Code

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a non-normal population
population = np.random.exponential(scale=2, size=10000)

sample_means = []

# Draw 1000 samples of size 30
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.hist(sample_means, bins=30, density=True, color="skyblue", edgecolor="black")

plt.title("Central Limit Theorem")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)

plt.show()
```

## Applications
- Statistical inference
- Confidence intervals
- Hypothesis testing
- Machine Learning
- Data Science
- Quality control

---

## Key Points
- Works for **any population distribution**.
- As **sample size increases**, the sample mean becomes **normally distributed**.
- Larger sample size gives a **better approximation** to the normal distribution.
- The standard deviation of sample means is called the **Standard Error**.

---

## Quick Example

Suppose exam marks are **not normally distributed**.

- Take **1000 random samples**.
- Each sample contains **30 students**.
- Compute the **average marks** of each sample.
- Plot these averages.

➡️ The graph of the sample means will look **approximately normal**, even though the original marks are not normally distributed.

---

## Revision Tips
- **Population** → Original data
- **Sample** → Small subset of the population
- **Sample Mean** → Average of a sample
- **CLT** → Sample means follow a **Normal Distribution** when **n ≥ 30**
- **Standard Error** = **σ / √n**

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate a non-normal population
population = np.random.exponential(scale=2, size=10000)

sample_means = []

# Draw 1000 samples of size 30
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot the distribution of sample means
plt.hist(sample_means, bins=30, density=True, color="skyblue", edgecolor="black")

plt.title("Central Limit Theorem")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)

plt.show()
````
